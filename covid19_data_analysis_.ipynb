```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

### importing covid19 dataset
 


```python
corona_dataset_csv = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
corona_dataset_csv.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
corona_dataset_csv.head()
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

### I have found a very interesting result in this analysis, this result shows that people who are living in developed countries are more prone to getting the corona virus infection as compared to less developed countries, possible reasons can be-


1.   Developed countries usually have older populations, and age is one of the key factors from spreading and dying from covid19. Thus, developed countries are more prone to the infection.
2.   As developed countries have more resources and better medical infrastructure, a much larger number of test were performed, therefore number of detected cases are higher in developed countries.
